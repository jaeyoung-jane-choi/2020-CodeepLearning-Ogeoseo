```python
#basic libraries 

import pandas as pd 
import numpy as np


import seaborn as sns
sns.set_palette("colorblind")
from matplotlib import pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(20,5))


import os
import datetime
import warnings
warnings.filterwarnings("ignore")

os.chdir("/Users/heeyoon/Documents/::2020_2/coDeep/DATA")

print(os.getcwd())

data = pd.read_csv('1113total.csv')
print(data.shape) #(662402, 13)
print(data.columns)
data.isnull().sum()
```

```python
## id 재수정 => 아직 소수점인 id 존재

data['student_id'] = data['student_id']*10
data['student_id'] = data['student_id'].astype('int')
data['student_id'] = data['student_id'].astype('str')

data
```

```python
## 1. Make student_id and major table
nan_major = data[data['major'].isnull() == True]
major = data[data['major'].isnull() == False]

print(nan_major.shape) #(158835, 13)
print(major.shape)     #(503567, 13)

sid_major = major.drop_duplicates(['student_id'], keep='first')
sid_major = sid_major[['student_id', 'major']]
sid_major.reset_index(inplace = True, drop = True)
sid_major
```

```python
## 2. fill Nan major using sid-major table
def fillmajor(row):
    try:
        major = sid_major[sid_major['student_id'] == row['student_id']].iloc[:,1].values[0]
    except:    
        major = np.nan  
    return major

nan_major['major'] = nan_major.apply(fillmajor, axis=1, result_type = 'expand')

print(nan_major.isnull().sum()) # major 13163
```

```python
## 3. concat major and filled nan_major
total = pd.concat([major, nan_major])
print(total.shape) #(662402, 13)

## 4. two version - with / without major
total_w_major = total[total['major'].isnull() == False]
total_wo_major = total[total.columns.difference(['major'])]

print(total_w_major.shape)  #(649239, 13)
print(total_wo_major.shape) #(662402, 12)
```

```python
### select book location = “취업자료”, “과제도서" “미디어 자료실", “비도서 자료", “학위논문”

##### 1차 : 도서위치로 삭제 
total_wo_major2 = total_wo_major[~total_wo_major['book_loc_2'].isin(['학위논문(제2서고)', '4층 과제도서', '수험서 코너',
                                                                     '과제도서(Service+ Center)','과제도서', 
                                                                     '4층 취업자료코너','3층 미디어자료실', '미디어자료실', 
                                                                     '일반자료실취업서가(3층)','건축학 비도서'])]

total_wo_major2 #639950 rows × 12 columns
print(total_wo_major2.isnull().sum()) # chungu_num = 4


##### 2차 : 4개 남음
total_wo_major2 = total_wo_major2[total_wo_major2['chungu_num'].isnull() == False]  
print(total_wo_major2.shape) # (639946, 12)
```

```python
### 대출 건수로 자르기
      
borrow_cnt = total_wo_major2.groupby('student_id').count()
borrow_cnt = borrow_cnt[['borrow_date_time']]
borrow_cnt.sort_values('borrow_date_time', inplace = True)
id_over5 = borrow_cnt[borrow_cnt['borrow_date_time']>= 5]

final_data = total_wo_major2[total_wo_major2['student_id'].isin(id_over5.index)]
print(final_data.shape) #(615075, 12)
```

```python
final_data.rename(columns = {'seoji_num':'meta_id', 
                             'real_return_date_2': 'r_date',
                             'dungrok_num' : 'book_id',
                             'chungu_num': 'ddc',
                             'borrow_date_time' : 'b_date'
                            }, inplace=True)


final_data2 = final_data[['student_id', 'book_id', 'ddc', 'title', 'meta_id', 'borrow_type','b_date', 'r_date', 'return_type']]
final_data2.sort_values('b_date', inplace = True)
final_data2.reset_index(inplace = True, drop = True)
final_data2['meta_id'] = final_data2['meta_id'].astype('str')

final_data2 #615075 rows × 9 columns
```

```python
final_data2.to_csv('1114_data.csv', index = 0)
final_data2.info()
```
